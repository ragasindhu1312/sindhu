```python
#include<stdio.h>
int ispalindrome(int n)
{
	//numbers----yes----true----1
	//numbers----no----false----0
	int rev=0,buffer=n;
	while(n!=0)
	{
		rev=rev*10+(n%10);
		n/=10;//n=n//10;
	}
	if(buffer==rev)
	{
		return 1;
	}
	else
	{
		return 0;
	}
}
int main()
{
	int i,a1,a2,res,pcnt=0;
	scanf("%d%d",&a1,&a2);
	for(i=a1;i<=a2;i++)
	{
		if(i%2==0)
		{
			res=ispalindrome(i);
			if(res==1){
				pcnt++;
			}
		}
	}
	printf("%d",pcnt);
}
```

```python
#include<stdio.h>
int factorial(int x)
{
	int i,fact=1;
	for(i=x;i>0;i--)
	{
		fact=fact*i;
	}	
	return fact;
}
int main()
{
	int n,r,s=0,buffer;
	scanf("%d",&n);
	buffer = n;
	while(n!=0)
	{
		r=n%10;
		s=s+factorial(r);
		n=n/10;
	}
	if(s==buffer)
	{
		printf("yes");
	}
	else
	{
		printf("no");
	}
	return 0;
}
```

```python
#include<stdio.h>
int factors(int i)
{
    int a,s=0;
    for(a=1;a<i;a++)
    {
     if(i%a==0)
     {
         s=s+a;
     }
    }
    return s;
}


int main()
{
    int a1,a2,i,b=0;
    scanf("%d""%d",&a1,&a2);
    for(i=a1;i<=a2;i++)
    {
      if(i==factors(i))  
      {
          b++;
          printf("%d\n",i);
         
      }
     
    }
     printf("the no.of perfect no.'s in the given range are %d",b);
     
    return 0;
}
```

```python
#include<stdio.h>
#include<math.h>
int digitcount(int n)
{
	int cnt=0;
    while (n!=0)
	{
		cnt++;
	    n/=10;	
	}	
	return cnt;
}
int main()
{
	int x,r,sum=0,dgtcnt;
	scanf("%d",&x);
	dgtcnt=digitcount(x);
	while(x!=0)
	{
	r=x%10;
	sum=sum+pow(r,dgtcnt);
	x=x/10;
	}
	printf("%d",sum);
	return 0;
}
```
